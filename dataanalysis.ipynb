```python
# Create a Machine Learning Model
# y = Price
# x = Anything other than Price
# We have to create a Regression Model

from sklearn.preprocessing import LabelEncoder
```

```python
df
```

```python
lb = LabelEncoder()
feature = {}
for k in df.columns:
    if df[k].dtypes=='O':
        df[str(k)+'_Encode'] = lb.fit_transform(df[k])
        gt = {}
        ke = df[[k,str(k)+'_Encode']].drop_duplicates()[str(k)+'_Encode'].values
        val = df[[k,str(k)+'_Encode']].drop_duplicates()[k].values
        for i in range(len(ke)):
            gt[val[i]] = str(ke[i])
        feature[k] = gt
```

```python
feature
```

```python
df
```

```python
# Create JSON file to store the feature data for multiple calling

import json
with open('encdata.json','w') as jsonfile:
    json.dump(feature,jsonfile,indent=4)
```

```python
df.drop(['Make','Model','Fuel','Registration city','Car documents','Assembly','Transmission','Condition'],axis=1,inplace = True)
```

```python
df
```

```python
# Input data
x = df[['Year',"KM's driven",'Make_Encode','Model_Encode','Fuel_Encode','Registration city_Encode','Car documents_Encode','Assembly_Encode','Transmission_Encode']]
```

```python
# Output data
y = df['Price']
```

```python
# Split the data into training and testing dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=2529)
```

```python
x_train
```

```python
x_test
```

```python
# Import your Machine Learning Algorithm
from sklearn.linear_model import LinearRegression
```

```python
# Create your ML Model Object
lr = LinearRegression()
```

```python
# Fit the training dataset into Machine Learning
lr.fit(x_train,y_train)
```

```python
# Now its ready for prediction
lr.predict([[2013,80000,2,41,3,0,1,0,0]])
```

```python
lr.predict([[2011,126544,9,52,3,25,1,1,1]])
```

```python
# Checking the Performance using X_test & y_test Data set
lr.score(x_test,y_test)
```

```python
# Creating our Model File
import pickle
with open('model.pickle','wb') as model_file:
    pickle.dump(lr,model_file)
```
